# Consistência

Existem colunas com valores "Null" que são string, a seguir as colunas e a quantidade desses valores será identificada.

```python
df_operacao['bairro'][6]
```

Abaixo temos a quantidade total desses registros por Coluna. 

```python
df_nulls = pd.DataFrame([(coluna, len(df_operacao[df_operacao[coluna]=="Null"][coluna])) for coluna in df_operacao.columns], columns = ['Coluna', 'Qtd "Null"'])
df_nulls['% Qtd "Null"'] = (df_nulls['Qtd "Null"']/df_operacao.shape[0])*100
df_nulls.sort_values(by='Qtd "Null"', ascending=False)
```
